# Criando meu tokenizer

```python
import re
import urllib.request

def pre_tokenizer(text):
  result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
  result = [item for item in result if item.strip()]

  return result
```

```python
#Testando em shakespeare:

url = "https://www.gutenberg.org/files/100/100-0.txt"
file_path = "shakespeare.txt"

urllib.request.urlretrieve(url, file_path)

with open(file_path, "r", encoding="utf-8") as f:
    text = f.read()

result = pre_tokenizer(text)
vocabulary = list(set(result))
vocabulary.extend(["<UNK>", "<PAD>"])
vocab_size = len(vocabulary)

print(vocabulary)
print(vocab_size)
```

```python
#Criando tokens para cada palavra do vocabulário:

tokens = {token:integer for integer,token in enumerate(vocabulary)}
tokens
```

```python
#Criando meu Tokenizer

class meu_tokenizer:
  def __init__(self, vocabulary):
    self.str_to_int = tokens
    self.int_to_str = {i:s for s,i in tokens.items()}

  def encode(self, text):
    result = re.split(r'([,.:;?_!"()\']|--|\s)', text)
    result = [item for item in result if item.strip()]
    ids = [self.str_to_int[s] for s in result]
    return ids

  def decode(self, ids):
    tokens = [self.int_to_str[i] for i in ids]
    text = " ".join(tokens)
    return text

meu_tokenizer = meu_tokenizer(tokens)
```

# Criando o dataloader

```python
import torch
from torch.utils.data import Dataset, DataLoader


class GPTDatasetV1(Dataset):
    def __init__(self, txt, tokenizer, max_length, stride):
        self.input_ids = []
        self.target_ids = []

        # Tokenize the entire text
        token_ids = tokenizer.encode(txt, allowed_special={"<|endoftext|>"})

        # Use a sliding window to chunk the book into overlapping sequences of max_length
        for i in range(0, len(token_ids) - max_length, stride):
            input_chunk = token_ids[i:i + max_length]
            target_chunk = token_ids[i + 1: i + max_length + 1]
            self.input_ids.append(torch.tensor(input_chunk))
            self.target_ids.append(torch.tensor(target_chunk))

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return self.input_ids[idx], self.target_ids[idx]
```

```python
def create_dataloader_v1(txt, batch_size, max_length, stride=1,
                         shuffle=True, drop_last=True, num_workers=0):
    # Initialize the tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")

    # Create dataset
    dataset = GPTDatasetV1(txt, tokenizer, max_length, stride)

    # Create dataloader
    dataloader = DataLoader(
        dataset, batch_size=batch_size, shuffle=shuffle, drop_last=drop_last, num_workers=num_workers)

    return dataloader
```

# Função para testar o tokenizer

```python
import urllib.request
import tiktoken
import re

def analyze_text_from_url(titulo, url):
    headers = {
         "User-Agent": "Mozilla/5.0"
    }
    req = urllib.request.Request(url, headers=headers)
    response = urllib.request.urlopen(req)
    text = response.read().decode("utf-8")

    # tokenizer
    tokenizer = tiktoken.get_encoding("gpt2")
    encoded = tokenizer.encode(text)

    words = text.split()
    num_tokens = len(encoded)
    num_words = len(words)
    tokens_per_word = num_tokens / num_words if num_words > 0 else 0

    print(f'Título do livro: {titulo}')
    print(f'Alguns tokens: {encoded[140:155]}')
    print(f'Quantidade de palavras: {num_words}')
    print(f'Quantidade de tokens: {num_tokens}')
    print(f'Quantidade de tokens por palavra {tokens_per_word:.4f}')
```
